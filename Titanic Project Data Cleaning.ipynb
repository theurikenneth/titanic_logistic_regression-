```python
# conversion from categorical to numeric values
train = pd.read_csv('/home/theuri/after_removal_outliers.csv')
```

```python
train.head()
```

```python
pd.get_dummies(train['Sex'], prefix='Sex')
```

```python
Sex = pd.get_dummies(train['Sex'], prefix = 'Sex')
Embarked = pd.get_dummies(train['Embarked'], prefix = 'Embarked')
Pclass = pd.get_dummies(train['Pclass'], prefix = 'Plcass')
```

```python
new_train = pd.concat([train,Sex,Embarked, Pclass], axis=1)
new_train
```

```python
#deleting extra data
new_train.drop(['PassengerId', 'Pclass', 'Name', 'Sex', 'Embarked', 'Ticket', 'Cabin'], axis = 1, inplace = True)
new_train.head()
```

```python
new_train.to_csv('final_dataset.csv', index = False)
```
